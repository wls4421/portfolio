```python
#-*- coding: utf-8 -*- 
import selenium
import os
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from time import sleep
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
```

```python
# make DataFrame
Hotel_Info_df = pd.DataFrame(columns = ['name', 'location', 'facilities'])
Reviews_df = pd.DataFrame(columns = ['name', 'reviewerID', 'review_HeadLine', 'review_MainText', 'grade'])
global result
#Avoid_Dup = pd.read_csv('C:/Users/wls44/Desktop/workspace (4)/Hotel_Info.csv' )
##크롤링
# 페이지 맨 밑으로 휠 내리기
def FullScrolling() :
    for c in range(0,50):
        driver.find_element_by_tag_name('body').send_keys(Keys.PAGE_DOWN)
        sleep(0.3)
def DeepScrolling():
     for c in range(0,75):
        driver.find_element_by_tag_name('body').send_keys(Keys.PAGE_DOWN)
        sleep(0.5)
def QuickScrolling() :
     for c in range(0,35):
        driver.find_element_by_tag_name('body').send_keys(Keys.PAGE_DOWN)
        sleep(0.1)
def ScrollUP() :
    driver.find_element_by_tag_name('body').send_keys(Keys.PAGE_UP)
    
def PopUpClose():
    try:
        PopUp = driver.find_element_by_xpath('/html/body/div[36]/div/div/div[2]/span')
        PopUp.click()
    except :
        print("Nothing")
def FindFacilities(a, text):
    facilities_css = list()
    try:
        facilities_css = driver.find_elements_by_css_selector('#abouthotel-features > div.Box-sc-kv6pi1-0.cTxLvk.FeatureGroup > div:nth-child({0}) > ul > li'.format(a))
    except:
        print("FindFacilities can't find css")
        return None
    facilities_text = []
    for i in range(0,len(facilities_css)) :
        temp = facilities_css[i].get_attribute('innerText')
        facilities_text.append(temp)
    if text in facilities_text :
        result = text
    else :
        print("%s cannot find"%text)
    if len(text) > 2 :
        return result
    else :
        print("Non Text")
def resetPage():
    driver.close()
    driver.switch_to.window(driver.window_handles[0])
#Initial List
HotelName_list = []
reviewerID_list = []
reviewHeadLine_list = []
reviewMainText_list = []
location_list = []
grade_list = []
```

```python
def GetReview(text):
    global Hotel_Info_df
    global Reviews_df
    '''global Avoid_Dup
    copy = Avoid_Dup[Avoid_Dup['name'] == text]
    copy2 = Hotel_Info_df[Hotel_Info_df['name'] == text]
    if len(copy) != 0:
        print("중복발견")
        HotelName_list = None
        reviewerID_list = None
        reviewHeadLine_list = None 
        reviewMainText_list = None
        grade_list = None
        return HotelName_list, reviewerID_list, reviewHeadLine_list, reviewMainText_list, grade_list
    elif len(copy2) != 0:
        print("중복발견")
        HotelName_list = None
        reviewerID_list = None
        reviewHeadLine_list = None 
        reviewMainText_list = None
        grade_list = None
        return HotelName_list, reviewerID_list, reviewHeadLine_list, reviewMainText_list, grade_list
    else:
        print("중복없음")'''
    
    
    count = 1
    HotelName_list = []
    reviewerID_list = []
    reviewHeadLine_list = []
    reviewMainText_list = []
    facilities_list = []
    grade_list = []
    try:
        DeepScrolling()
    except Exception as e:    
            print('Scrolling has been error.', e)
    try:
        locationCopy = driver.find_element_by_xpath('//*[@id="abouthotel-panel"]/div[3]/div/div[1]/div[1]/div/div/div[2]/span').get_attribute('innerText')
    except:
        try:
            locationCopy = driver.find_element_by_xpath('//*[@id="property-main-content"]/div[2]/div[2]/div[2]/span[1]').get_attribute('innerText')
        except:
            print("Location NULL")
            locationCopy = "NULL"
    try:
        facilities_list.append(FindFacilities(3,"무료 Wi-Fi (모든 객실)"))
        facilities_list.append(FindFacilities(4,"피트니스 센터"))
        facilities_list.append(FindFacilities(6,"레스토랑"))
        facilities_list.append(FindFacilities(7,"금연 숙소"))
        facilities_list.append(FindFacilities(10,"주차장(무료)"))
    except :    
            print('facility append has been error.')
    try:
        for i in range(0, len(facilities_list)):
            if(facilities_list[i] == None):
                del facilities_list[i]
                i -= 1
    except:    
            print('facility search has been error.')
    Hotel_Info_df = Hotel_Info_df.append({'name':text,
                                                    'location':locationCopy,
                                                    'facilities':facilities_list},
                                                    ignore_index=True)
    while count < 52 :
        count += 1
        try:
            for k in range(0,20):
                try:
                    reviewerID_list.append(driver.find_element_by_xpath('//*[@id="review-{0}"]/div[1]/div/div[2]/strong'.format(k)).get_attribute('innerText'))
                    HotelName_list.append(text)
                    reviewHeadLine_list.append(driver.find_element_by_xpath('//*[@id="review-{0}"]/div[2]/div[1]/div[1]/p[1]'.format(k)).get_attribute('innerText'))
                    reviewMainText_list.append(driver.find_element_by_xpath('//*[@id="review-{0}"]/div[2]/div[1]/div[1]/p[2]'.format(k)).get_attribute('innerText'))
                    grade_list.append(driver.find_element_by_xpath('//*[@id="review-{0}"]/div[1]/div/div[1]/div[1]'.format(k)).get_attribute('innerText'))
                except Exception as e:   
                    continue
            QuickScrolling()
            ScrollUP()
            ScrollUP()
            sleep(0.3)
            try:
                button = driver.find_element_by_css_selector('#reviewSection > div:nth-child(6) > div > span:nth-child(3)')
                button.click()
            except:
                try:
                    QuickScrolling()
                    ScrollUP()
                    ScrollUP()
                    button.click()
                except:
                    break
            sleep(1)
        except:
            break
    return HotelName_list, reviewerID_list, reviewHeadLine_list, reviewMainText_list, grade_list
```

```python
# 페이지에 효시된 h3 tag(호텔제목) 긁어서 리스트화
def MakePageName_list() :
    agodaName = []
    agodaName = (driver.find_elements_by_tag_name('h3'))
    Name_list = []
    Href_list = []
    delNum = -1
    print("agodaName 갯수",len(agodaName))
    for i in range(0, len(agodaName)):
        Name_list.append(agodaName[i].get_attribute('innerText'))
        
    Name_list = set(Name_list)
    Name_list = list(Name_list)
    
    for i in range(0, len(Name_list)):
        try:
             Href_list.append(driver.find_element_by_partial_link_text(Name_list[i]))
        except:
            continue
    print(Name_list)
    print("Name_list 갯수",len(Name_list))
    print("Href_list 갯수",len(Href_list))
    driver.execute_script("window.scrollTo(0, 0)")
    
    return len(Href_list), Href_list, Name_list
```

```python
def Whole() :
    global Hotel_Info_df
    global Reviews_df
    sleep(3)
    PopUpClose()
    FullScrolling()
    i, Href, Name_list = MakePageName_list()
    for t in range(0,i):
        try:
            PopUpClose()
            try:
                Href[t].click()
                sleep(3)
            except:
                continue
        except Exception as e:    
            print('Href.click has been error.', e)
        driver.switch_to.window(driver.window_handles[1])
        HotelName_list, reviewerID_list, reviewHeadLine_list, reviewMainText_list, grade_list = GetReview(Name_list[t])
        if(HotelName_list!= None):
            for p in range(0, len(HotelName_list)):
                try:
                    Reviews_df = Reviews_df.append({'name':HotelName_list[p],
                                                        'reviewerID':reviewerID_list[p],
                                                        'review_HeadLine':reviewHeadLine_list[p],
                                                        'review_MainText':reviewMainText_list[p],
                                                        'grade':grade_list[p]},
                                                        ignore_index=True)
                except:    
                    continue
        driver.close()
        sleep(3)
        driver.switch_to.window(driver.window_handles[0])
        sleep(3)
                
def ToNextPage():
        try:
            driver.find_element_by_xpath('//*[@id="paginationNext"]').send_keys(Keys.ENTER)
        except Exception as e:   
            print('ToNextPage has been error.', e)
```

```python
while True:
    try:
        Whole()
        ToNextPage()
    except:
        break;
```
